```python
import matplotlib.image as mpimg 
from scipy import misc
from scipy import ndimage
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
spec = mpimg.imread(r"\Users\Kaustav Dutta\Desktop\spectrogram1.png");
plt.imshow(spec),plt.title('Raw Spectrogram')
plt.show()
```

```python
# Convert color image to grayscale to help extraction of edges

gray = np.dot(spec, [1,1,1])
plt.imshow(gray,cmap = plt.get_cmap('gray'));
```

```python
#Define a threhold. If the pixel is above threshold, mark it as 1, otherwise 0.
T=1.17
```

```python
for i in range(1, gray.shape[0]-1):
    for j in range(1, gray.shape[1]-1):
        if (gray[i,j])< T:
            gray[i,j]=0
        else:
            gray[i,j]=1
```

```python
plt.imshow(gray),plt.title('Binary Raw Spectrogram')
plt.show()
```

```python
img=cv2.imread(r"\Users\Kaustav Dutta\Desktop\binary1.png")
gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
edges=cv2.Canny(gray,50,150,apertureSize=3)
lines=cv2.HoughLinesP(edges,2,np.pi/180,100,minLineLength=0,maxLineGap=150)
print(lines)
```

```python
for line in lines:
    x1,y1,x2,y2=line[0]
    cv2.line(img,(x1,y1),(x2,y2),(0,0,250),2)
```

```python
cv2.imshow('image',img)
k=cv2.waitKey(0)
cv2.destroyAllWindows()
```
